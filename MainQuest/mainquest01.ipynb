```python
import random as r

class Account:
    bankname = 'SC은행'
    n_account = 0 #계좌 수 초기화
    big_balance_customer = []

    def __init__(self, name, first_balance): #초기 입금주와 초기 잔액 입력 받고 랜덤 계좌 생성
        self.name = name #고객이름 받기
        self.balance = first_balance #초기금액 받기
        self.account_number = self.make_account_number()  #랜덤계좌 생성
        Account.n_account += 1 #생성된 계좌수를 세는 역할
        self.number_deposit = 0 #입금횟수
        self.deposit_history_list = [] #입금기록
        self.withdraw_history_list = [] #출금기록

    def get_account_num(self): #생성된 계좌수 출력
        print(self.n_account)

    def deposit(self): #입금함수 최소 1원 이상
        while True: #무한반복
            d = int(input('입금할 금액을 입력하세요 : '))

            if d > 0: #최소 입금액이 1원 이상 그리고 5회 입금하면 이자 추가
                self.balance += d
                self.number_deposit += 1
                self.deposit_history_list.append(str(d))
                if self.number_deposit == 5:
                    self.balance *= 1.01
            else: #입금금액이 1원 미만일 경우 출력
                print('입금 금액은 최소 1원 이상만 가능합니다.')
                continue

            while True: #입금을 더 할건지 안할건지 물어보는 반복문
                check_more = int(input('입금을 더 하시겠습니까?(YES : 1, NO : 2) : '))
                if check_more == 1:
                    break
                elif check_more ==2:
                    break
                else:
                    print('입력은 1,2만 할 수 있습니다.')
            if check_more == 1:
                continue
            else:
                break

    def withdraw(self): #출금함수
        while True: #무한반복
            w = int(input('출금할 금액을 입력하세요 : '))

            if w < self.balance: #최대 출금액이 잔고를 넘지 못함
                self.balance -= w
                self.withdraw_history_list.append(str(-w))

            else: #최대 출금액을 넘어갈 경우
                print('출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.')
                continue

            while True: #출금을 더 할건지 안할건지 물어보는 반복문
                check_more = int(input('출금을 더 하시겠습니까?(YES : 1, NO : 2) : '))
                if check_more == 1:
                    break
                elif check_more ==2:
                    break
                else:
                    print('입력은 1,2만 할 수 있습니다.')
            if check_more == 1:
                continue
            else:
                break


    def make_account_number(self): #000-00-000000 랜덤 계좌 생성
        account_number = str(r.randint(0,9))+str(r.randint(0,9))+str(r.randint(0,9))+ '-' +str(r.randint(0,9))+str(r.randint(0,9))+'-'+ str(r.randint(0,9))+str(r.randint(0,9))+str(r.randint(0,9))+str(r.randint(0,9))+str(r.randint(0,9))+str(r.randint(0,9))
        return account_number

    def display_info(self): #회원 정보 출력
        add = format(self.balance, ',')
        print(f'은행이름 : {self.bankname}, 예금주 : {self.name}, 계좌번호 : {self.account_number}, 잔고 : {add}')

    def check_balance(self): #잔액이 100만원 이상일 경우
        if self.balance >= 1000000:
            self.display_info()
    def deposit_history(self): #입금 기록 출력
        print(f'{self.name}의 입금 기록입니다. : {self.deposit_history_list}')
    def withdraw_history(self): #출금 기록 출력
        print(f'{self.name}의 출금 기록입니다. : {self.withdraw_history_list}')

a = Account("배인수", 1000000)
b = Account("문인재", 50000)
c = Account("문동재", 60000)

c_list = [a, b, c]

print(f'생성된 계좌 수 : {Account.n_account}')

for i in c_list:
    i.deposit()
    i.withdraw()
    i.display_info()
    i.deposit_history()
    i.withdraw_history()

print('계좌에 100만원 이상인 고객님 리스트')
for i in c_list:
    i.check_balance()
```
